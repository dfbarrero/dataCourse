## Matplotlib interfaces

Matplotlib comes with two different interfaces.

- Matlab-like. Old-fashioned function-oriented API. Provided by the *matplotlib.plt* module.

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)");
```

- Object-oriented. Object-oriented and more powerfull API.

```python
fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x))
ax.set_title("A sine curve")
ax.set_xlabel("x")
ax.set_ylabel("sin(x)");
```

Two concepts to know:
- Figure: The canvas, it can contain several axes.
-Axes: The plot (do not confuse with axis!).

## Multiple plots

Several methods to draw several plots.

```python
import matplotlib.pyplot as plt # plt provides Matlab API style

plt.figure()  # create a plot figure

# create the first of two panels and set current axis
plt.subplot(2, 1, 1) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis
plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x));
```

```python
# First create a grid of plots
# ax will be an array of two Axes objects
fig, ax = plt.subplots(2)

# Call plot() method on the appropriate object
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))

ax[0].set_title("sin(x)")
ax[0].set_ylabel("y")
ax[1].set_title("cos(x)")
ax[1].set_ylabel("y");
```

```python
fig = plt.figure(figsize =(5, 4)) 
  
# Creating first axes for the figure
ax1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) 
  
# Creating second axes for the figure
ax2 = fig.add_axes([0.5, 0.5, 0.3, 0.3])
  
# Adding the data to be plotted
ax1.plot(x, np.sin(x))
ax2.plot([1, 2, 3, 4, 5], [2, 3, 4, 5, 6])
plt.show()
```

Adjust lines

```python
plt.plot(x, x + 0, '-')  # solid default color
plt.plot(x, x + 1, '-g')  # solid green
plt.plot(x, x + 2, '--c') # dashed cyan
plt.plot(x, x + 3, '-.k') # dashdot black
plt.plot(x, x + 4, ':r');  # dotted red
```

```python
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
  
# Adding the legends
plt.legend(["sin(x)", "cos(x)"])
plt.show()
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.ylim(-2,2)

plt.legend();
```

## Three-dimensional plots

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
z = [1, 8, 27, 64, 125]
# Creating the figure object
fig = plt.figure()
# keeping the projection = 3d
# ctreates the 3d plot
ax = plt.axes(projection = '3d')
ax.plot3D(z, y, x);
```

```python
ax = plt.axes(projection = "3d")

# Data for a three−dimensional line
zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, "gray")

# Data for three−dimensional scattered points
zdata = 15 * np.random.random(100)
xdata = np.sin(zdata) + 0.1 * np.random.randn(100)
ydata = np.cos(zdata) + 0.1 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c = zdata, cmap = "Greens");
```

```python
from matplotlib import cm

n_radii = 8
n_angles = 36

# Make radii and angles spaces
radii = np.linspace(0.125, 1.0, n_radii)
angles = np.linspace(0, 2*np.pi, n_angles, endpoint=False)[..., np.newaxis]

# Convert polar (radii, angles) coords to cartesian (x, y) coords.
x = np.append(0, (radii*np.cos(angles)).flatten())
y = np.append(0, (radii*np.sin(angles)).flatten())
z = np.sin(-x*y)

# Plot
fig, ax = plt.subplots(subplot_kw={'projection': '3d'})
ax.plot_trisurf(x, y, z, vmin=z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[]);
```

```python
import matplotlib.pyplot as plt
from matplotlib import cm
import numpy as np

plt.style.use('_mpl-gallery')

# Make data
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

# Plot the surface
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Blues)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[]);
```

Prepare data.

```python
def f(x, y):
    return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)

x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

3-D visualizations:

```python
plt.contour(X, Y, Z, colors='black');
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```
